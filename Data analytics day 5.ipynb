```python
q1 = np.percentile(df['salary'], 25)
q3 = np.percentile(df['salary'], 75)
```

```python
q1, q3
```

```python
iqr = q3 - q1
```

```python
upper_fence = q3 + 1.5 * iqr
```

```python
lower_fence = q1 - 1.5 * iqr
```

```python
upper_fence, lower_fence
```

```python
df
```

```python
df_without_outliers = df[ (df['salary'] < upper_fence) & (df['salary'] > lower_fence) ]
df_without_outliers
```
