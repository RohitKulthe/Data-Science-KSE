```python
students =[56,65,46,51,61,45]

courses = ['java','python','webd','kotlin','Dart','Docker']
```

```python
#pie chart
```

```python
plt.title('pie chart')
plt.pie(students, labels = courses,autopct ='%.2f%%',shadow=True,radius=0.9,labeldistance=1.2);
```

```python
plt.title('Pie chart')
colors = ['red','blue','cyan','grey','skyblue']
startangle= 30
explode = [0,0,0.2,0,0,0]
plt.pie(students, labels = courses,autopct ='%.2f%%',shadow=True,radius=0.9,labeldistance=1.2, startangle = startangle, colors=colors, explode=explode
       )
plt.savefig('MY PLOT');
```
