# Normalization

```python
x = np.array([11, 45, 23, 17, 21, 15, 7, 33])
```

```python
def normalize(a):
    norm = (a-a.min())/(a.max()-a.min())
    return norm
```

```python
normalize(x)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title( 'Actual Data')
plt.plot(x)
plt.subplot(1,2,2)
plt.title( 'Normalize Data' )
plt.plot(normalize(x))
```

```python
df = pd.read_csv('Social_Network_Ads (1).csv')
df
```

```python
y = df['Purchased']
x = df[['Age', 'EstimatedSalary']]
```

```python
plt.xlabel('age')
plt.ylabel('salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c = y)
```

```python
x_scaled
```

```python
type(x_scaled)
```

```python
x_new = pd.DataFrame(x_scaled, columns=x.columns)
```

```python
x_new
```

```python
x_new.describe()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual Data')
plt.xlabel('age')
plt.ylabel('salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c = y)

plt.subplot(1,2,2)
plt.title('Normalized Data')
plt.xlabel('age')
plt.ylabel('salary')
plt.grid()
plt.scatter(x_new['Age'], x_new['EstimatedSalary'], c = y)
```
